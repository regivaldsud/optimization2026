# Problem Statement

Solve the Cylinder Problem considering the following data:

* N: 10
* $c_1$: 2
* $c_2$: 0.5
  

```julia
# codeusing Pkg
Pkg.add("Optim")
```

```julia
using Optim

N  = 10.0
c1 = 2.0
c2 = 0.5

# função objetivo a maximizar
zfun(x) = begin
    r, h = x
    N*pi*h*r^2 - c1*pi*r^2 - c2*pi*h*(pi*r^2 + 2*pi*r*h)*r^2
end

# Optim minimiza, então usamos o negativo
f(x) = -zfun(x)

lower = [1e-8, 1e-8]
upper = [10.0, 10.0]
x0    = [1.0, 1.0]

res = optimize(f, lower, upper, x0, Fminbox(NelderMead()))

xopt = Optim.minimizer(res)
r_opt, h_opt = xopt
z_opt = zfun(xopt)

println("r ótimo = ", r_opt)
println("h ótimo = ", h_opt)
println("z ótimo = ", z_opt)
```

# Problem Statement

Solve the 3-bus Optimal Power Flow Problem with following data:



```julia
using Pkg
Pkg.add("JuMP")
Pkg.add("Ipopt")
```

```julia
using JuMP, Ipopt

# -----------------------------
# Dados do sistema
# -----------------------------
y = 1 / (0.01 + 0.1im)
g = real(y)
b = imag(y)

# Matriz Ybus
G = [
    2g   -g   -g
    -g   2g   -g
    -g   -g   2g
]

B = [
    2b   -b   -b
    -b   2b   -b
    -b   -b   2b
]

Pd = [0.0, 0.0, 0.8]
Qd = [0.0, 0.0, 0.6]

# linhas: (i, j, Smax)
lines = [
    (1, 2, 0.25),
    (1, 3, 2.0),
    (2, 3, 2.0)
]

# -----------------------------
# Modelo
# -----------------------------
model = Model(Ipopt.Optimizer)
set_silent(model)

# Geração ativa
@variable(model, 0 <= Pg[1:3])
set_upper_bound(Pg[1], 3.0)
set_upper_bound(Pg[2], 0.8)
fix(Pg[3], 0.0; force = true)

# Geração reativa
@variable(model, -2 <= Qg[1:3] <= 2)
fix(Qg[3], 0.0; force = true)

# Magnitudes de tensão
@variable(model, 0.95 <= V[1:3] <= 1.10)

# Ângulos
@variable(model, δ[1:3])
fix(δ[3], 0.0; force = true)

# Chute inicial
set_start_value.(Pg, [0.1, 0.7, 0.0])
set_start_value.(Qg, [0.3, 0.3, 0.0])
set_start_value.(V,  [1.05, 1.05, 1.00])
set_start_value.(δ,  [0.01, 0.02, 0.0])

# Função objetivo: min 2*p1 + 1*p2
@objective(model, Min, 2*Pg[1] + Pg[2])

# -----------------------------
# Balanço de potência AC
# -----------------------------
for i in 1:3
    @NLconstraint(model,
        Pg[i] - Pd[i] ==
        sum(V[i]*V[j]*(G[i,j]*cos(δ[i]-δ[j]) + B[i,j]*sin(δ[i]-δ[j])) for j in 1:3)
    )

    @NLconstraint(model,
        Qg[i] - Qd[i] ==
        sum(V[i]*V[j]*(G[i,j]*sin(δ[i]-δ[j]) - B[i,j]*cos(δ[i]-δ[j])) for j in 1:3)
    )
end

# -----------------------------
# Limites de fluxo nas linhas
# -----------------------------
for (i, j, smax) in lines
    # fluxo i -> j
    @NLconstraint(model,
        (
            g*V[i]^2 - V[i]*V[j]*(g*cos(δ[i]-δ[j]) + b*sin(δ[i]-δ[j]))
        )^2
        +
        (
            -b*V[i]^2 - V[i]*V[j]*(g*sin(δ[i]-δ[j]) - b*cos(δ[i]-δ[j]))
        )^2
        <= smax^2
    )

    # fluxo j -> i
    @NLconstraint(model,
        (
            g*V[j]^2 - V[i]*V[j]*(g*cos(δ[j]-δ[i]) + b*sin(δ[j]-δ[i]))
        )^2
        +
        (
            -b*V[j]^2 - V[i]*V[j]*(g*sin(δ[j]-δ[i]) - b*cos(δ[j]-δ[i]))
        )^2
        <= smax^2
    )
end

# -----------------------------
# Resolver
# -----------------------------
optimize!(model)

println("Status = ", termination_status(model))
println("Custo ótimo = ", objective_value(model))
println()

println("Pg = ", value.(Pg))
println("Qg = ", value.(Qg))
println("V  = ", value.(V))
println("δ  = ", value.(δ))

# -----------------------------
# Função auxiliar para fluxos
# -----------------------------
function line_flow(i, j, Vv, δv, g, b)
    Pij = g*Vv[i]^2 - Vv[i]*Vv[j]*(g*cos(δv[i]-δv[j]) + b*sin(δv[i]-δv[j]))
    Qij = -b*Vv[i]^2 - Vv[i]*Vv[j]*(g*sin(δv[i]-δv[j]) - b*cos(δv[i]-δv[j]))
    Sij = sqrt(Pij^2 + Qij^2)
    return Pij, Qij, Sij
end

Vv = value.(V)
δv = value.(δ)

println("\nFluxos nas linhas:")
for (i, j, smax) in lines
    Pij, Qij, Sij = line_flow(i, j, Vv, δv, g, b)
    Pji, Qji, Sji = line_flow(j, i, Vv, δv, g, b)
    println("Linha $i -> $j : P = $(Pij), Q = $(Qij), |S| = $(Sij), limite = $(smax)")
    println("Linha $j -> $i : P = $(Pji), Q = $(Qji), |S| = $(Sji), limite = $(smax)")
end
```

# Problem Statement

A small engineering consulting firm has 3 senior designers available to work on the firm's 4 current projects over the next 2 weeks. Each designer has 80 hours to split among the projects, and the following table shows the manager's scoring $(0=$ nil to $100=$ perfect $)$ of the capability of each designer to contribute to each project, along with his estimate of the hours that each project will require.


|      Designer     | Project 1    | Project 2    | Project 3    | Project 4    |
|----------|--------------|--------------|--------------|--------------|
| 1        | 90           | 80           | 10           | 50           |
| 2        | 60           | 70           | 50           | 65           |
| 3        | 70           | 40           | 80           | 85           |



|     **Required:**      | Project 1    | Project 2    | Project 3    | Project 4    |
|-----------|--------------|--------------|--------------|--------------|
| **Hours** | 70           | 50           | 85           | 35           |


## Model

Let the design engineers be set $E$ with $E_{i} \; :i \in [1,2,3]$ and the projects be $P$ with  $P_{j} \; :j \in [1,2,3,4]$. We can model the problem as allocation of the number of hours $H_{ij}$ with each design engineer $E_{i}$ $\forall i $ that are being put onto the projects $P_{j}$ $\forall j $, given the $i^{th}$ engineer $E$ works on $j^{th}$ project with given score $e_{ij}$.

Let the maximum hours available with each engineer be $H_{max}$ and the required number of hours for each project $P_j$ be $R_{j} \; \forall j $. Thus, then the mathematical formulation can be made as:

$$
\text{Maximize } \sum_{i \in E} \sum_{j \in P} H_{ij} e_{ij}
$$

Subject to:

$$
\sum_{j \in P} H_{ij} \leq H_{max} \;\; \forall i \in E
$$

$$
\sum_{i \in E} H_{ij} \geq R_j \;\; \forall j \in P
$$

$$
H_{ij} \geq 0 \;\; \forall i,j
$$


## Implement

```julia
# codeusing Pkg
Pkg.add("JuMP")
Pkg.add("HiGHS")
```

```julia
using JuMP
using HiGHS

# Matriz de pontuações
score = [
    90 80 10 50;
    60 70 50 65;
    70 40 80 85
]

# Horas requeridas por projeto
req = [70, 50, 85, 35]

# Máximo de horas por designer
max_hours = 80

n_designers = size(score, 1)
n_projects  = size(score, 2)

model = Model(HiGHS.Optimizer)

# x[i,j] = horas do designer i no projeto j
@variable(model, x[1:n_designers, 1:n_projects] >= 0)

# Função objetivo
@objective(model, Max, sum(score[i, j] * x[i, j] for i in 1:n_designers, j in 1:n_projects))

# Restrição de horas de cada designer
@constraint(model, [i in 1:n_designers], sum(x[i, j] for j in 1:n_projects) <= max_hours)

# Restrição de horas requeridas por projeto
@constraint(model, [j in 1:n_projects], sum(x[i, j] for i in 1:n_designers) >= req[j])

optimize!(model)

println("Status = ", termination_status(model))
println("Valor ótimo = ", objective_value(model))
println()

println("Alocação ótima de horas:")
for i in 1:n_designers
    for j in 1:n_projects
        println("Designer ", i, " -> Projeto ", j, " = ", value(x[i, j]))
    end
end

println()
println("Horas usadas por designer:")
for i in 1:n_designers
    println("Designer ", i, ": ", sum(value(x[i, j]) for j in 1:n_projects))
end

println()
println("Horas recebidas por projeto:")
for j in 1:n_projects
    println("Projeto ", j, ": ", sum(value(x[i, j]) for i in 1:n_designers))
end
```

# Problem Statement

A dietitian is planning a meal that meets the daily nutritional requirements for calories, protein, and vitamins at a minimum cost.


|      Food Item     | Cost ($)    | Calories    | Protein (g)    | Vitamins (% Daily)    |
|----------|--------------|--------------|--------------|--------------|
| Apple        | 1              | 100           | 0.5           | 2            |
| Bread        | 0.50           | 200           | 4             | 0            |
| Milk         | 2              | 150           | 8             | 10           |
| Egg          | 0.30           | 70            | 6             | 0            |

Daily nutritional requirements: 500 calories, 50g protein, 100% vitamins.

Define decision variables: $y_1$ for Apples, $y_2$ for Bread, $y_3$ for Milk, $y_4$ for Eggs.  
$y_i$ represents the quantity of each food item.

$$
\begin{aligned}
\text{Minimize} \quad & y_1 + 0.5y_2 + 2y_3 + 0.3y_4 \\
\text{Subject to} \quad 
& 100y_1 + 200y_2 + 150y_3 + 70y_4 \geq 500 \\
& 0.5y_1 + 4y_2 + 8y_3 + 6y_4 \geq 50 \\
& 2y_1 + 0y_2 + 10y_3 + 0y_4 \geq 100 \\
& y_1, y_2, y_3, y_4 \geq 0
\end{aligned}
$$

Ensure all dietary requirements for calories, protein, and vitamins are met.

```julia
# codeusing Pkg
Pkg.add("JuMP")
Pkg.add("HiGHS")

using JuMP, HiGHS

model = Model(HiGHS.Optimizer)

# Variáveis
@variable(model, y1 >= 0)   # Apple
@variable(model, y2 >= 0)   # Bread
@variable(model, y3 >= 0)   # Milk
@variable(model, y4 >= 0)   # Egg

# Função objetivo: minimizar custo
@objective(model, Min, 1.0*y1 + 0.5*y2 + 2.0*y3 + 0.3*y4)

# Restrições
@constraint(model, 100*y1 + 200*y2 + 150*y3 + 70*y4 >= 500)   # calorias
@constraint(model, 0.5*y1 + 4*y2 + 8*y3 + 6*y4 >= 50)         # proteína
@constraint(model, 2*y1 + 10*y3 >= 100)                       # vitaminas

optimize!(model)

println("Status = ", termination_status(model))
println("Custo mínimo = ", objective_value(model))
println("y1 (Apple) = ", value(y1))
println("y2 (Bread) = ", value(y2))
println("y3 (Milk)  = ", value(y3))
println("y4 (Egg)   = ", value(y4))
```

# Problem Statement

Consider a hiker who needs to choose the most valuable items for a hike without overloading the backpack.

- Items: Tent (Value: $\$120$, Weight: 2kg), Stove (Value: $\$80$, Weight: 1kg), Food (Value: $\$60$, Weight: 1kg)  
- Backpack capacity: 3.5kg  

Objective: Maximize the value of items in the backpack.

Define binary decision variables: $x_1$ for Tent, $x_2$ for Stove, $x_3$ for Food.  
$x_i = 1$ if the item is chosen, and 0 otherwise.

$$
\begin{aligned}
\text{Maximize} \quad & 120x_1 + 80x_2 + 60x_3 \\
\text{Subject to} \quad 
& 2x_1 + x_2 + x_3 \leq 3.5 \\
& x_1, x_2, x_3 \in \{0,1\}
\end{aligned}
$$

```julia
# codeusing Pkg
Pkg.add("JuMP")
Pkg.add("HiGHS")

using JuMP, HiGHS

model = Model(HiGHS.Optimizer)

# Variáveis binárias
@variable(model, x1, Bin)   # Tent
@variable(model, x2, Bin)   # Stove
@variable(model, x3, Bin)   # Food

# Função objetivo
@objective(model, Max, 120x1 + 80x2 + 60x3)

# Restrição de capacidade
@constraint(model, 2x1 + x2 + x3 <= 3.5)

optimize!(model)

println("Status = ", termination_status(model))
println("Valor máximo = ", objective_value(model))
println("x1 (Tent)  = ", value(x1))
println("x2 (Stove) = ", value(x2))
println("x3 (Food)  = ", value(x3))
```
