## Chapter 13: Quasi-Newton methods

### Algorithm 13.1: Quasi-Newton BFGS method





Example 5.8: $f(x_1,x_2) = \frac{1}{2} x_1^2 + x_1 \cos(x_2)$

```julia
f(x) = 0.5 * x[1]^2 + x[1] * cos(x[2])
grad(x) = [x[1] + cos(x[2]), -x[1] * sin(x[2])]
x, hist, status = bfgs(f, grad, [1.0, 1.0], eps=1e-8)
println("status = ", status)
println("ponto = ", fmt(x))
println("f(x) = ", fmt(f(x)))
println("||grad|| = ", fmt(norm(grad(x))))
println("iteracoes = ", length(hist) - 1)
```
